```python
%sql
USE CATALOG dev;

CREATE SCHEMA IF NOT EXISTS olist_bronze;

CREATE SCHEMA IF NOT EXISTS olist_silver;

CREATE SCHEMA IF NOT EXISTS olist_gold;
```

```python
%sql
SHOW SCHEMAS;
```

```python
customers_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("dev.olist_bronze.olist_customers")
```

```python
%sql
SELECT * FROM dev.olist_bronze.olist_customers
LIMIT 10;
```

```python
bronze_tables = {
    "olist_customers": customers_df,
    "olist_orders": orders_df,
    "olist_order_items": order_items_df,
    "olist_products": products_df,
    "olist_order_payments": payments_df,
    "olist_order_reviews": reviews_df,
    "olist_sellers": sellers_df,
    "olist_geolocation": geolocation_df,
    "product_category_translation": categories_df
}
```

```python
for table_name, df in bronze_tables.items():

    print(f"Creating {table_name}...")

    (
        df.write
          .format("delta")
          .mode("overwrite")
          .saveAsTable(f"dev.olist_bronze.{table_name}")
    )

print("All Bronze tables created successfully!")
```
